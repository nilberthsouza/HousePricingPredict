```python
train_index = train.shape[0]
test_index = test.shape[0]

Id = test['Id']

y_train = train['SalePrice'].copy
train.drop(['SalePrice'],axis=1,inplace=True)

full = pd.concat(objs=[train,test],axis=0).reset_index(drop=True)
```

```python
#full.drop(['PoolQC','MiscFeature','Alley','Fence'],axis=1,inplace=True)

fireplacequ = full['FireplaceQu'].mode(dropna=True)
full['FireplaceQu'].fillna(fireplacequ, inplace=True)

lotfrontage = full['LotFrontage'].mode(dropna=True)
full['LotFrontage'].fillna(lotfrontage, inplace=True)

garageyrblt = full['GarageYrBlt'].mode(dropna=True)
full['GarageYrBlt'].fillna(garageyrblt, inplace=True)

garagetype = full['GarageType'].mode(dropna=True)
full['GarageType'].fillna(garagetype, inplace=True)

garagefinish = full['GarageFinish'].mode(dropna=True)
full['GarageFinish'].fillna(garagefinish, inplace=True)

garagecond = full['GarageCond'].mode(dropna=True)
full['GarageCond'].fillna(garagecond, inplace=True)

garagequal = full['GarageQual'].mode(dropna=True)
full['GarageQual'].fillna(garagequal, inplace=True)

bsmtcond = full['BsmtCond'].mode(dropna=True)
full['BsmtCond'].fillna(bsmtcond, inplace=True)

bsmtexposure = full['BsmtExposure'].mode(dropna=True)
full['BsmtExposure'].fillna(bsmtexposure, inplace=True)



```
